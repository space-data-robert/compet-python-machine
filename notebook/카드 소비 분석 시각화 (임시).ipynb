```python
data_tmp = data[data.amount > 0].groupby(
    ['year_month', 'year']
).amount.count().reset_index()

figure = px.bar(
    data_tmp,
    x='year_month', y='amount', color='year',
    text_auto=True, width=800, height=300
)
figure.update_layout(
    bargap=0.2, font=dict(size=12), 
    legend=dict(title='거래년도'),
    margin=dict(t=20, l=20, b=20, r=20)
)
figure.update_xaxes(title_text='거래일자')
figure.update_yaxes(title_text='카드취소비율')
```

```python
data_tmp = data[data.amount < 0].groupby(
    ['year_month', 'year']
).amount.count().reset_index()

figure = px.bar(
    data_tmp,
    x='year_month', y='amount', color='year',
    text_auto=True, width=800, height=300
)
figure.update_layout(
    bargap=0.2, font=dict(size=12), 
    legend=dict(title='거래년도'),
    margin=dict(t=20, l=20, b=20, r=20)
)
figure.update_xaxes(title_text='거래일자')
figure.update_yaxes(title_text='카드취소량')
```
